# Building ML Predictive Model

```python
# Scale Features
from sklearn.preprocessing import StandardScaler
df.ffill(inplace=True)

if df.shape == 1: 
        pass
else:
        scaler = StandardScaler()
        scaled_data = scaler.fit_transform(df)

# Step 2: Feature Engineering

# Feature Selection
selected_features = df.drop(['Exited'], axis=1)
target_variable = df['Exited']
scaled_data.shape
```

```python
# Step 3: Model Selection and Training

# Split data into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(selected_features, target_variable, test_size=0.2, random_state=42)

# Train and Evaluate Models
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score

# Random Forest Classifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
```

```python
new = pd.read_csv(r'C:\Users\shame\Desktop\bank\data\cc.csv')
rf_model.predict(new)
```

```python
# Calculate accuracy
rf_accuracy = accuracy_score(y_test, rf_predictions)
# Calculate Precision
precision = precision_score(y_test, rf_predictions)
# Calculate Recall
recall = recall_score(y_test, rf_predictions)
# Calculate F1-score
f1 = f1_score(y_test, rf_predictions)

# Print the results

print("Accuracy score:",rf_accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
